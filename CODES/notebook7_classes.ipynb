# Classes and Object Oriented Programming in Python

A class defines the behaviour of an object and the kind of information an object can store. The information in a class is stored in attributes, and functions that belong to a class are called methods. A child class inherits the attributes and methods from its parent class. 

```python
# Creating a Dog class 
class Dog():
    """Represents a dog class"""
    def __init__(self,name):
        """Initialize dog object."""
        self.name = name
    def sit(self):
        """Similuate sitting"""
        print(self.name + " is sitting")

my_dog = Dog('Bruno')

print(my_dog.name + " is a great dog!")
my_dog.sit()
```

```python
# Inheritance
class SARDog(Dog):
    """Represents a search Dig."""

    def __init__(self,name):
        """Initialize the sardog"""
        super().__init__(name)

    def search(self):
        """Simulate searching"""
        print(self.name + " is searching")

my_sar_dog = SARDog('Willie')

print(my_dog.name + " is a search dog.")
my_dog.sit()
my_dog.search() # This will raise an AttributeError since my_dog is not a SARDog
```

```python
my_sar_dog.sit()
my_sar_dog.search()
```
