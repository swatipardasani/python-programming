```python
# CG : 

## Question : Write a Python function to read a large CSV file in chunks and calculate the sum of a numeric column 
# called amount without loading the entire file into memory.

import pandas as pd

def sum_amount_in_chunks(file_path, chunk_size=100000):
    """
    Reads a large CSV file in chunks and calculates the sum of the 'amount' column.
    
    Parameters:
        file_path (str): Path to the CSV file.
        chunk_size (int): Number of rows per chunk (default = 100,000).
    
    Returns:
        float: Total sum of the 'amount' column.
    """
    total_sum = 0
    
    # Read file in chunks
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        # Add the sum of 'amount' column from this chunk
        total_sum += chunk['amount'].sum()
    
    return total_sum

# Example usage:
# total = sum_amount_in_chunks("large_sales_data.csv", chunk_size=50000)
# print("Total sum of amount:", total)



```
