In Python, pickling means converting a Python object into a byte stream (serialization), while unpickling means converting that byte stream back into a Python object (deserialization). 

🟦 Pickling

Definition: Serialization of Python objects into a binary format.
    wb - write binary

Purpose: Store objects in files, databases, or transmit them over a network.

Syntax:

```python
import pickle

data = {"name": "Alice", "age": 25, "skills": ["Python", "SQL"]}

# Pickling: write object to file
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)
```

🟦 Unpickling

Definition: Deserialization — loading the binary data back into Python objects.
 rb - read binary

Syntax:

```python
import pickle

# Unpickling: read object from file
with open("data.pkl", "rb") as f:
    loaded_data = pickle.load(f)

print(loaded_data)
# Output: {'name': 'Alice', 'age': 25, 'skills': ['Python', 'SQL']}
```

<pre>
🟦 Why Binary Mode?
Binary mode is used when working with non‑text files such as:

Pickled objects (pickle.dump / pickle.load)

Images (.jpg, .png)

Audio/video files

Any format where data isn’t plain text
